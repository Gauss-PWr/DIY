## Transform

```python
df.dtypes
```

```python
import numpy as np

def nanint(x):
    
    try:
        return int(x)
    except ValueError:
        return np.nan


df['Released_Year'] = df['Released_Year'].apply(nanint)
```

```python
df['Runtime'] = df.Runtime.apply(lambda x: int(x.replace(' min', '')))
```

```python
def nangross(x):
    
    try:
        return int(x.replace(',', ''))
    except AttributeError:
        return np.nan

df['Gross'] = df.Gross.apply(nangross)
```
